```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
# slecting the datatype for multilinear regression
df_new = df.select_dtypes(include=['int64','float64'])
df_new

```

```python
x = df_new.iloc[:,0:3].values
x
y = df_new.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred #predicted output
```

```python
y_test #actual output
```

```python
#pandas plotting
df1=pd.DataFrame({'actual':y_test,'predicted':y_pred})
df1
```

```python
df1.plot(figsize=(20,9))
```

```python
df1.plot(figsize=(20,9),kind = 'bar')
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
